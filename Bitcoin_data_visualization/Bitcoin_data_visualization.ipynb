```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

bitcoin_data = pd.read_csv('btc-usd-max.csv')
```

```python
bitcoin_data
```

```python
bitcoin_data.info()
```

```python
bitcoin_data['snapped_at'] = pd.to_datetime(bitcoin_data['snapped_at'])
```

```python
bitcoin_data
```

```python
plt.figure(figsize=(10, 6))
plt.plot(bitcoin_data['snapped_at'], bitcoin_data['price'], linestyle='-', color='b')

# Add labels and title
plt.title('Bitcoin Price Over Time', fontsize=16)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Price', fontsize=12)

# Rotate x-axis labels for better readability (optional)
# plt.xticks(rotation=45)

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
